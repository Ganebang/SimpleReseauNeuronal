```python
#Import des packages
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.models import load_model
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
import matplotlib.pyplot as plt
import numpy as np
import argparse
import random
import pickle
import cv2
import os
```

```python
# Chemin d'accès aux images
my_path ='./dataset/'
```

```python
def load_data(data_dir, new_size=None):
    if not new_size is None:
        img_rows, img_cols = new_size
    labels=[]
    data =[]
    for label in  os.listdir(data_dir):
        for img_path in os.listdir(data_dir+label+'/'):
            img = cv2.imread(data_dir+label+'/'+img_path)
            img=cv2.resize(img, (img_rows, img_cols)).flatten()
            lab = img_path.split('_')[0]
            data.append(img)
            labels.append(lab)
    return data, labels
```

```python
data, labels = load_data(my_path, (32, 32))
```

```python
# Mettre les intensités brutes de chaque pixel dans [0, 1]
data = np.array(data, dtype="float") / 255.0
labels = np.array(labels)
```

```python
# Séparer notre data en trainSet et testSet
(trainX, testX, trainY, testY) = train_test_split(data, labels, test_size=0.25, random_state=42)
```

```python
# Convertir les labels d'entiers en vecteurs par exemple cats = [1 0 0], dogs = [0 1 0], panda = [0 0 1] 
labelBin = LabelBinarizer()
```

```python
# fit_transform() choisit des labels uniques pour l'entrainement
trainY = labelBin.fit_transform(trainY)
testY = labelBin.transform(testY)
```

```python
# Définition d'une architecture 3072-1024-512-3 à l'aide keras
classifier = Sequential()
classifier.add(Dense(1024, input_shape=(3072, ), activation="sigmoid"))
classifier.add(Dense(512, activation="sigmoid"))
classifier.add(Dense(len(labelBin.classes_), activation="softmax"))
```

```python
# Initialiser notre taux d'apprentissage de départ et les époques d'entrainemeent
INIT_LR = 0.01
EPOCHS = 100
```

```python
# Compilation du modèle à l'aide SGD (Stochastic Gradient Descent)
print("[INFO] entrainement du modèle ...")
optimizer = SGD(lr=INIT_LR)
classifier.compile(loss="categorical_crossentropy", optimizer=optimizer, metrics=["accuracy"])
# Entraîner le réseau neuronal
H = classifier.fit(x=trainX, y=trainY, validation_data=(testX, testY), epochs=EPOCHS, batch_size=32)
# Evaluer le réseau neuronal
print("[INFO] évaluation du modèle ...")
predictions = classifier.predict(x=testX, batch_size=32)
print(classification_report(testY.argmax(axis=1), predictions.argmax(axis=1), target_names=labelBin.classes_))
```

```python
# Tracer la précision et la perte de l'entraînement
N = np.arange(0, EPOCHS)
plt.figure()
plt.plot(N, H.history["loss"], label="train_loss")
plt.plot(N, H.history["val_loss"], label="val_loss")
plt.plot(N, H.history["accuracy"], label="train_acc")
plt.plot(N, H.history["val_accuracy"], label="val_acc")
plt.title("Training Loss and Accuracy (Simple RN)")
plt.xlabel("Epoch #")
plt.ylabel("Loss/Accuracy")
plt.legend()
plt.show()
```

```python
# Sauvegarder notre modèle et les labels bineariser sur le disque
classifier.save('./output/SRN_classifier', save_format="h5")
f = open('./output/SRN_label.pickle', "wb")
f.write(pickle.dumps(labelBin))
f.close()
```

```python
# Charger le modèle et le convertisseur binaire des labels
classifier = load_model('./output/SRN_classifier')
labelBin = pickle.loads(open('./output/SRN_label.pickle', "rb").read())
```

```python
# Charger une image et la classifier
image = cv2.imread('./test/chat2.jpg')
output = image.copy()
plt.figure(figsize = (10,10))
plt.imshow(image)
plt.show()
```

```python
image = cv2.resize(image, (32, 32))
image = image.flatten()
image = image.astype("float") / 255.0
image = image.reshape((1, image.shape[0]))
```

```python
# Faire la prédiction
predicts = classifier.predict(image)
```

```python
# Trouver la classe 
i = predicts.argmax(axis=1)[0]
label = labelBin.classes_[i]
```

```python
# Représenter l'image et sa classe prédite
text = "{}: {:.2f}%".format(label, predicts[0][i] * 100)
cv2.putText(output, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 3)

plt.figure(figsize = (10,10))
plt.imshow(output)
plt.show()
```
